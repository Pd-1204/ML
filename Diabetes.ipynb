```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score, precision_score,recall_score
```

```python
data= pd.read_csv('C:/Users/PD/Downloads/diabetes/diabetes.csv')
```

```python
# replace 0 values with mean values
for column in data.columns[1:-3]:
    data[column].replace(0,np.NaN, inplace=True)
    
data.head(10)
```

```python
data.describe()
```

```python
X= data.iloc[:,:8]
Y= data.iloc[:,8:]
```

```python
x_train, x_test, y_train, y_test= train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
knn=KNeighborsClassifier(n_neighbors=11)
knn_fit= knn.fit(x_train, y_train.values.ravel())
knn_pred=knn_fit.predict(x_test)
print(knn)
```

```python
print("confusion matrix", confusion_matrix(y_test, knn_pred))
print("accuracy score", accuracy_score(y_test, knn_pred))
print("recall score", recall_score(y_test, knn_pred))
print("precision score", precision_score(y_test, knn_pred))
print("error rate", 1-accuracy_score(y_test, knn_pred))
```
